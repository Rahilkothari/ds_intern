# Task # 4 - To Explore Decision Tree Algorithm

For the given ‘Iris’ dataset, create the Decision Tree classifier and
visualize it graphically. The purpose is if we feed any new data to this
classifier, it would be able to predict the right class accordingly.

```python
##Importing all the necessary libraries
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Iris.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
#EDA(Exploratory Data Analysis)
sns.pairplot(df,palette = 'Set1')
```

```python
#Splitting the data
from sklearn.model_selection import train_test_split

X = df.drop('Species', axis = 1)
y = df['Species']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
#Model
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier()
```

```python
#Fitting the data
tree.fit(X_train,y_train)
```

```python
#Prediction
y_pred = tree.predict(X_test)
y_pred
```

```python
#Evaluation
from sklearn.metrics import accuracy_score
```

```python
#Classification report
print(accuracy_score(y_test,y_pred))
```

```python
#Loading the same dataset for the purpose of using it directly for visualization
import sklearn.datasets as datasets
import pandas as pd

iris=datasets.load_iris()

#Dataframe
df=pd.DataFrame(iris.data, columns=iris.feature_names)
print(df.head(5))

y=iris.target
```

```python
#Model - Decision Tree
from sklearn.tree import DecisionTreeClassifier
tree=DecisionTreeClassifier()
tree.fit(df,y)
```

```python
#Visualizing the tree
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
```

```python
dot_data = StringIO()
export_graphviz(tree, out_file=dot_data, feature_names=iris.feature_names,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```

```python
#Predicting new data
tree.predict([[4.1,3.7,2.5,0.5]])
```
